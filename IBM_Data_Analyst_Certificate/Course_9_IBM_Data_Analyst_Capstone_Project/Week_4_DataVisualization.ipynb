## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp 
FROM master
"""
df = pd.read_sql_query(QUERY,conn)
df.plot(kind = "hist", figsize = (8, 6))
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

QUERY = """
SELECT Age 
FROM master
"""
# sns.set_palette('Dark2')
plt.figure(figsize = (6, 4))
df = pd.read_sql_query(QUERY,conn)
sns.boxplot(data = df, x = df['Age'])
plt.title('Age Distribution of Respondents')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs 
FROM master
"""
plt.figure(figsize = (10, 5))
df = pd.read_sql_query(QUERY,conn)
sns.scatterplot(data = df, x = df['Age'], y = df['WorkWeekHrs'])
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs 
FROM master
"""
plt.figure(figsize = (10, 7))
df = pd.read_sql_query(QUERY,conn)
sns.scatterplot(data = df, y = 'CodeRevHrs', x = 'WorkWeekHrs', size = 'Age', sizes=(10, 2000), legend = False, alpha = 0.35)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) AS Respondents
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY COUNT(DatabaseDesireNextYear) DESC
LIMIT 5
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.set_index('DatabaseDesireNextYear', inplace = True)
colors = ['#ff9ff3', '#feca57', '#ff6b6b', '#48dbfb', '#1dd1a1']
df['Respondents'].plot(kind = 'pie',
                 figsize=(15, 6),
                 autopct='%1.1f%%', 
                 startangle=90,
                 shadow=True,
                 colors = colors, 
                 labels = None,
                 pctdistance=1.2,
                 fontsize = 14)
plt.legend(labels=df.index, loc='upper right')
plt.ylabel('Percentage of Respondents', fontsize = 14)
plt.title('Top 5 Databases That Respondents Wish To Learn', fontsize = 16, y = 1.1) 
plt.axis('equal') 
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age Between 30 and 35
"""
# the read_sql_query runs the sql query and returns the data as a dataframe

df = pd.read_sql_query(QUERY,conn)
df1 = df.groupby('Age').median()
df1.plot(kind = 'bar', stacked = True, figsize = (8, 5)) 
plt.title('Median WorkWeekHrs And CodeRevHrs For The Age Group 30 To 35')
plt.legend(loc = 'right')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT  ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60

"""
# the read_sql_query runs the sql query and returns the data as a dataframe
df2 = pd.read_sql_query(QUERY,conn)
df2.groupby('Age').median().plot(kind = 'line', figsize = (15, 6), fontsize = 12)
plt.title('Median ConvertedComp For Ages From 45 To 60', fontsize = 16)
plt.xlabel('Age', fontsize = 16)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY = """
SELECT MainBranch, Count(*) as Count
FROM master
group by MainBranch

"""
# the read_sql_query runs the sql query and returns the data as a dataframe
df3 = pd.read_sql_query(QUERY,conn)
df3.plot(kind = 'barh', figsize = (8, 6))
plt.title('MainBranch')
plt.ylabel('Developers / Non-Developers by Profession')
plt.show()
```
